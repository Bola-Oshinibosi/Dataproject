```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# checking for duplicate. There is no duplicate
bank.duplicated().sum()
```

```python
# CLEANING DATA
# CHNAGE ATTRITION_FLAG TO 1 AND 0

bank["Attrition_Flag"] = bank["Attrition_Flag"].str.replace("Attrited Customer","1")
bank["Attrition_Flag"] = bank["Attrition_Flag"].str.replace("Existing Customer","0")
```

```python
bank["Gender"] = bank["Gender"].str.replace("M","Male")
```

```python
bank["Gender"] = bank["Gender"].str.replace("F","Female")
```

```python
# Changing Attrition_flag data type integer

bank["Attrition_Flag"] = pd.to_numeric(bank.Attrition_Flag, errors='coerce')
```
